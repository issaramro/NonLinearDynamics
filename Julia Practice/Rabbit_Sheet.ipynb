```julia
using CairoMakie
```

```julia
# we have the solutions and the form of equs
# treating a problem as a map
# 4D plot
# we need (x,y) as the origin of the vector and (xdot, ydot) for direction

xs = 0:0.1:4
ys = 0:0.1:4

x = zeros((length(xs), length(ys)))
y = zeros((length(xs), length(ys)))
u = zeros((length(xs), length(ys)))
v = zeros((length(xs), length(ys)))

# enumerate return the index and the element
for (i, _x) ∈ enumerate(xs), (j, _y) ∈ enumerate(ys)
    x[i,j] = _x
    y[i,j] = _y
    u[i,j] = _x*(3-_x -2_y)
    v[i,j] = _y*(2-_x - _y)
end
# make them vectors
x = reshape(x,length(xs)*length(ys));
y = reshape(y,length(xs)*length(ys));
u = reshape(u,length(xs)*length(ys));
v = reshape(v,length(xs)*length(ys));

# we now want to worry about directions only, we want vectors smaller than separation between points so that they dont overlap
mag = sqrt.(u.^2 + v.^2);
u = u./mag./15;
v = v./mag./15;
```

```julia
fig = Figure()
ax = Axis(fig[1,1])
arrows!(ax,x,y,u,v, color = mag, colormap=:viridis)
Colorbar(fig[1,2])
y1 = (3 .- xs)./2
y2 = 2 .- xs
lines!(ax, xs, y1)
lines!(ax, xs, y2)
xlims!(-0.1,3)
ylims!(-0.1,3)
# Intersection of x = 0 and y = 0: fixed points
fig
```
