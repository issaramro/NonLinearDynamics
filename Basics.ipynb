```julia
a = true
b = false
a && b # Worth noting that this is short circuit and will only evaluate b if a is true
a || b # same here but will only evaluate b if a is false
!a
```

### Tuples

Tuples are immutable, meaning you can't change them after they are created. They are created with `()`. You can access elements with `[]` and the index of the element. You can also use `end` to get the last element. You can also use `:` to get a range of elements. For example, `t[1:2]` will get the first two elements of the tuple.

```julia
tup = (1, "a", true,π)
println(tup[1])
println(tup[2:3])
tup[2:end]
```

### Named Tuples and Dictionaries

Named tuples are kind of like dictionaries, but they are immutable.


```julia
nt = (a = 1, b = 2, c = 3)
println(nt.a)
println(nt[1])
println(nt[:a])
```

Dicts are mutable, meaning you can change them after they are created. The are just like the `Python` dicts.

```julia
Contacts = Dict("John" => "555-555-5555", "Jane" => "555-555-5556")
println(Contacts["John"])
println(Contacts["Jane"])
Contacts["Jason"] = "555-555-5557"
println(Contacts["Jason"])
```

Functions can change the parameters that are passed to them.

By convention, functions that change their parameters are suffixed with `!`.

```julia
function add3!(x)
    x[1] += 3
    x
end

a = [1, 2, 3]
add3!(a)
a
```

Whereas functions that don't change the parameters (not `!`), would work like this

```julia
function add3(x)
    y = copy(x)
    y[1] += 3
    y
end

a = [1, 2, 3]
add3(a)
a
```

## Performance

Make sure that the type of the variables you are using is the same everywhere. For example, if you are using an array of floats, make sure that all the variables you are using are floats. If you are using an array of integers, make sure that all the variables you are using are integers. If you don't do this, Julia will have to convert the types of the variables every time you use them, which will slow down your code.

```julia
# Pkg.add("BenchmarkTools")
using BenchmarkTools
```

```julia
function mymax(init, x)
    s = init
    for xi in x
        s = s < xi ? xi : s
    end
    return s
end

# benchmark it

r = rand(Int, 1000);
@btime mymax(-Inf32, $r);
```

```julia
@btime mymax(-Inf, $r);
```

```julia
@btime mymax(typemin(Int), $r);
```

```julia
@code_warntype mymax(-Inf32, r)
```

```julia
@code_warntype mymax(typemin(Int), r)
```
