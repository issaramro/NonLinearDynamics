```julia
function f(r, x)
    r .+ x - log.(1 .+ x)
end

function df(x)
    1 - 1 / (1 .+ x)
end


function Newton(f, df, x0, ϵ, r)
    count = 0
    ro = x0
    o = 0
    while abs(ro - o) >= ϵ 
        o = ro
        ro= ro - (f(r,ro) / df(ro))
        count+=1
        if count >= 20
            return "no convergence"
            break
        end
    end
    return ro, count
end

x = range(-0.999,10, length = 150)
r = -2
print(Newton(f,df, 3,-0.5, r))


fig = Figure()
ax = Axis(fig[1,1])
lines!(x, f(r,x))
fig
```
