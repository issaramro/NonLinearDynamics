## Problem 3.3.2 - Maxwell-Bloch equations.
The Maxwell-Bloch equations are given by
$$\dot E = \kappa(P-E)$$
$$\dot P = \gamma_1 (ED - P)$$
$$\dot D = \gamma_2 (\lambda + 1- D - \lambda EP)$$
Where the parameters ($\kappa, \gamma_1, \gamma_2$) are positive and $\lambda$ may be positive, negative, or zero.<br>
a.) Studying the case where $\gamma_1, \gamma_2 >> \kappa$, D and P converge rapidly to steady behavior giving us $\dot D \approx 0, \dot P \approx 0$. We set $\dot P$ and $\dot D$ to zero in the second and third equations in the system, and we solve for P and D in terms of E. We get
$$P = ED$$ 
$$D = \frac{\lambda + 1}{\lambda E^2 + 1}$$
Plugging the results in the first equation of the system, we reduce the system to a one first order equation in E 
$$\dot E = \kappa E \left(\frac{\lambda + 1}{\lambda E^2 + 1} - 1 \right)$$
b.) To find the fixed points of the equation, we set $\dot E = 0$, getting 
$$E^*_1 = 0$$
$$E^*_2 = 1$$
$$E^*_3 = -1$$
These fixed points are independent of $\lambda$ but their stability depends on the sign of $\lambda$. We can see that by plotting the graph of $\dot E$ as function of $E$ and studying the stability.

```python
def MaxBloch_1d(E, k, l):
    return k*E*((l+1)/(l*E**2 + 1) - 1)

E = np.arange(-100, 100, 0.01)
k = 10
l = [-0.5, 0, 0.5]


plt.plot(E, MaxBloch_1d(E, k, l[0]), label = 'λ = -0.5')
plt.annotate('',xy=(-0.8,0),xytext=(-1,0),arrowprops={})
plt.annotate('',xy=(-1.2,0),xytext=(-1,0),arrowprops={})
plt.annotate('',xy=(0,0),xytext=(-0.2,0),arrowprops={})
plt.annotate('',xy=(0,0),xytext=(0.2,0),arrowprops={})
plt.annotate('',xy=(0.8,0),xytext=(1,0),arrowprops={})
plt.annotate('',xy=(1.2,0),xytext=(1,0),arrowprops={})
plt.title('λ = -0.5')
plt.ylabel("Ė")
plt.xlabel('E')
plt.xlim(-1.2, 1.2)
plt.ylim(-5,5)
plt.grid()
plt.show()


plt.plot(E, MaxBloch_1d(E, k, l[2]))
plt.title('λ = 0.5')
plt.annotate('',xy=(-1,0),xytext=(-0.8,0),arrowprops={})
plt.annotate('',xy=(-1,0),xytext=(-1.2,0),arrowprops={})
plt.annotate('',xy=(-0.2,0),xytext=(0,0),arrowprops={})
plt.annotate('',xy=(0.2,0),xytext=(0,0),arrowprops={})
plt.annotate('',xy=(1,0),xytext=(0.8,0),arrowprops={})
plt.annotate('',xy=(1,0),xytext=(1.2,0),arrowprops={})
plt.ylabel("Ė")
plt.xlabel('E')
plt.xlim(-1.2, 1.2)
plt.ylim(-5,5)
plt.grid()
plt.show()


plt.plot(E, MaxBloch_1d(E, k, l[1]), label = 'λ = 0')
plt.title('λ = 0')
plt.ylabel("Ė")
plt.xlabel('E')
plt.xlim(-1.2, 1.2)
plt.ylim(-5,5)
plt.grid()
plt.show()
```

c.) We can now plot the graph of the $E^*$ values as function of the sign of $\lambda$, getting us the below bifuraction diagram where the dashed lines signify unstable branches.

```python
plt.plot(np.arange(-9, 1), np.ones(10), color = 'b', linestyle='dashed')
plt.plot(np.arange(0, 10), np.ones(10), color = 'b')
plt.plot(np.arange(-9, 1), -1*np.ones(10), color = 'b', linestyle='dashed')
plt.plot(np.arange(0, 10), -1*np.ones(10), color = 'b')
plt.plot(np.arange(-9, 1), np.zeros(10), color = 'b')
plt.plot(np.arange(0, 10), -1*np.zeros(10), color = 'b', linestyle='dashed')
plt.grid()
plt.ylabel("E*")
plt.xlabel("λ")
plt.title("Bifurcation diagram")
plt.show()
```
