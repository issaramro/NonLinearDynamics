## Problem 3.6.6 - Patterns in Fluids.

The Landau equation describes the supercritical case of the dynamics of the amplitude of the pattersn in fluids near the bifurcation, and is given by
$$\tau \dot A = \epsilon A - gA^3 $$
Where $\tau$ represents the typical time scale, $\epsilon$ measures the distance from the bifurcation, and g > 0.<br>
The subcritical case of this equation is given by 
$$\tau \dot A = \epsilon A - gA^3 - kA^5$$
Where g<0 and k>0.

a.) We set the $\dot A$ term in the Landau equation to 0 to find the fixed points, getting
$$\epsilon = gA^2$$
$$A^* = \pm \sqrt{\frac{\epsilon}{g}} \propto \epsilon^{0.5}$$
Which agrees with Dubois and Bege's prediction.

b.) We now deal with the subcritical case for g = 0 (tricritical case), which is the borederline case between the supercritical and subcritical cases. We again set $\dot A$ to 0 to find the fixed points.
$$\epsilon = k A^4$$
$$A^2 = \pm \sqrt{\frac{\epsilon}{k}}$$
$$A = \pm \sqrt{ \pm \sqrt{\frac{\epsilon}{k}}} \propto \epsilon ^{0.25}$$

c.) Introducing a slight imperfection parameter h to the subcritical equation, we get
$$\tau \dot A = h + \epsilon A - gA^3 - kA^5$$
and we look at the the graph of $A^*$ as function of $\epsilon$ for the 3 cases: g < 0, g = 0, and g > 0.

```python
def Aitta(A, h, epsilon, k, g):
    return h + epsilon*A - g*A**3 - k*A**5
A = np.linspace(-10,10,1000)
h = 0.3
k = 0.5
g_val = [-1,0,1]

for g in g_val:
    plt.plot(A, Aitta(A, h, 0.1, k, g), label = f"g={g:.1f}")
plt.grid()
plt.ylim(-2, 2)
plt.xlim(-3,3)
plt.legend()
plt.ylabel("f(A)")
plt.xlabel("A")
plt.show()
```
