## Problem 3.5.4 - Bead on a Horizontal wire.

<img src="3.5.4.png"> 

a.) To find Newton's equation of motion for this system, we account for the 2 forces taking place: Damping force ($- b\dot x$) and Hooke's force ($-k\Delta L$) where $\Delta L$ is the projection of the displacement from the equilibrium length of the spring ($L_0$).
The elongated length of the spring makes an angle $\theta$ with the horizontal. To find the projected displacement of the spring from $L_0$, we have to subtract the projected length of the relaxed spring from the displacement x, such that 
$$\Delta L = x - L_0 \cos \theta$$
But $\cos \theta = \frac{x}{\sqrt{h^2 + x^2}}$. Hence, 
$$\Delta L = x - \frac{xL_0}{\sqrt{h^2 + x^2}}$$
We finally get the equation of motion
$$m \ddot x = - b\dot x - k (x - \frac{xL_0}{\sqrt{h^2 + x^2}})$$

b.) To find the fixed points of the equation, we set the velocity $\dot x$ to zero. Getting 
$$-\frac{m}{b}\ddot x-\frac{k}{b}x(1 - \frac{L_0}{\sqrt{h^2 + x^2}}) = 0$$
Solving for x would return the 3 fixed points.

c.) Taking m to be = 0, the equation becomes 
$$\dot x = - \frac{k}{b} (x - \frac{xL_0}{\sqrt{h^2 + x^2}})$$
and the fixed points become
$$x^* = 0$$
$$x^* = \sqrt{L_0^2 - h^2}$$
$$x^* = -\sqrt{L_0^2 - h^2}$$

```python
def Bead(k, b, L0, h, x):
    return -(k/b)*x*(1-L0/np.sqrt(h**2 + x**2))

k = 2
b = 1
L0 = 20
h = 1
x = np.arange(-100, 100, 0.01)

plt.plot(x, Bead(k, b, L0, h, x))
plt.xlabel("x")
plt.ylabel("ẋ")
plt.grid()
plt.xlim(-30,30)
plt.annotate('',xy=(4,0),xytext=(0,0),arrowprops={})
plt.annotate('',xy=(-4,0),xytext=(0,0),arrowprops={})
plt.annotate('',xy=(19.9,0),xytext=(25,0),arrowprops={})
plt.annotate('',xy=(19.9,0),xytext=(15,0),arrowprops={})
plt.annotate('',xy=(-19.9,0),xytext=(-15,0),arrowprops={})
plt.annotate('',xy=(-19.9,0),xytext=(-25,0),arrowprops={})
plt.show()
```

Checking for positive and negative values for h such that $|h|\ge L_0$, we notice that even though the values of the fixed points changes, their stability does not switch.<br>
We start by 3 fixed points for $h<0$ such that the zero is unstable and the 2 symmetric are stable. Then they all converge to one fixed point for $h = L_0$. Getting a pitchfork bifurcation.

```python
h_vals = np.arange(-5, 11)
x_vals_p = []
x_vals_n = []
L0 = 10

def x_star(L0, h):
    return np.sqrt(L0**2 - h**2)

for h in h_vals:
    x_vals_p.append(x_star(L0, h))
    x_vals_n.append(-1*x_star(L0, h))

plt.plot(h_vals, x_vals_p, color = "maroon")
plt.plot(h_vals, x_vals_n, color = "gray")
plt.plot(h_vals, np.zeros(len(h_vals)), color = "k", linestyle='dashed')
plt.grid()
plt.xlabel("h")
plt.ylabel("x*")
plt.title("Birfurcation diagram")
plt.show()
```
