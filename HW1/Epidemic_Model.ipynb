## Problem 3.7.6 - Model of an Epidemic.

An epidemic is modeled by this system
$$\dot x = - kxy$$
$$\dot y = kxy - ly$$
$$\dot z = ly$$
Where x(t) represents the number of healthy people, y(t) represents the number of sick people, and z(t) represents the number of dead people.

a.) We want to show that the total population is alwyas constant, that is x + y + z = N. <br>
If we assume that N = x + y + z, the differentiating both sides with respect to time would give us 
$$\dot N = \dot x + \dot y + \dot z = 0$$
But that is true since
$$\dot x + \dot y + \dot z = -kxy + kxy - ly + ly = 0$$

b.) From the $\dot z$ equation, we get $y = \frac{\dot z}{l}$. We replace this y expression in the equation of $\dot x$, we get
$$\frac{dx}{dt} = - \frac{k}{l}x \frac{dz}{dt}$$
Getting rid of the dt term and separating the variables, we get
$$\frac{1}{x} dx = - \frac{k}{l} dz$$
We integrate both sides, noting that x(0) = x_0, and we get
$$ln(x) = - \frac{k}{l}z + x_0$$
Exponentiating both sides gives us
$$x(t) = x_0 e^{-\frac{k}{l}z(t)}$$

c.) We have $\dot z = ly = l [N - z - x] = l\left[N - z - x_0 e^{-\frac{k}{l}z(t)}\right]$.

d.) To nondimensionlize the found $\dot z$ expression, I will define the dimensionless groups $u = \frac{kz}{l}$, $\tau = k x_0 t$, $a = \frac{N}{x_0}$, and $b = \frac{l}{kx_0}$. Getting
$$\frac{d u}{d \tau} = a - bu - e^{-u}$$

e.) a = $\frac{N}{x_0}$, where x_0 is the initial value of x(t). We know that N = x + y + z where (x,y,z) represent positive fraction of the population number. Hence we can say that $x \le N$ for all values of x including $x_0$. Therefore $\frac{N}{x_0} \ge 1$.<br>
b = $\frac{1}{k x_0}$, but we know k is a positive finite parameter and $x_0 \ge 0$, hence $\frac{1}{kx_0} > 0$. 

f.) To find the fixed points of the dimensionless equation, we set the derivative to 0.

```python
def epi(u, a, b):
    return a - b*u - np.exp(-u)
u = np.arange(-10, 100, 0.1)
a = 1.5
b = 1.5
plt.plot(u, epi(u, a, b))

plt.annotate('',xy=(-1.9,0),xytext=(-1.189,0),arrowprops={})
plt.annotate('',xy=(-0.4,0),xytext=(-1.189,0),arrowprops={})
plt.annotate('',xy=(0.653,0),xytext=(1.4,0),arrowprops={})
plt.annotate('',xy=(0.653,0),xytext=(-0.2,0),arrowprops={})

plt.xlim(-6,9)
plt.grid()
plt.ylim(-7, 4)
plt.xlabel("u")
plt.ylabel("du")
plt.show()
```

For all value of a $\ge$ 1 and b >0, we get 2 roots where the negative one is unstable and the positive is stable.

g.) To study what happens at the maximum of $\dot u (t)$, we set its derivative to zero. We start by $u = \frac{k}{l}z$,
$$\dot u = \frac{k}{l} \dot z = ky  $$
$$\ddot u = \frac{k}{l} \ddot z = k \dot y = 0$$
Hence the maxima of $\dot u$, $\dot z$, and y occur at the same time.

h.) We find the derivative of u at 0
$$\dot u (0)= a - bu(0) - e^{-u(0)}$$
Assuming that at t = 0, u(0) = 0, 
$$\dot u(0) = a - 1 > 1 > 0$$
Hence u is increasing at t = 0.

We can visualize that by plotting $\dot u$ as function of u.

```python
def epi(u, a, b):
    return a - b*u - np.exp(-u)
u = np.arange(-10, 100, 0.1)
a = 1.5
b = 0.4
plt.plot(u, epi(u, a, b))
plt.grid()
plt.xlabel("u")
plt.xlim(-20, 20)
plt.ylabel("du")
plt.show()
```

u evolves to eventually reach a stable fixed which means that $\dot u$ would converge to zero, something we can see from the above graph.<br>
The time for $\dot u$ to reach its maximum is $t_{peak} > 0$.
