```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
birds = []
ar = np.zeros(100)
nb_of_birds = 0

for i in range(100):
    n = np.random.randint(0,100)
    print(n)
    if n != 0 and ar[n-1] == 1:
        ar[n-1] = 0
        ar[n] = 1

    elif n != 99 and ar[n+1] == 1:
        ar[n+1] = 0
        ar[n] = 1
        
    elif n != 0 and n != 99 and ar[n-1] == 1 and ar[n+1] == 1:
        ar[n-1] = 0
        ar[n+1] = 0
        ar[n] = 1
        nb_of_birds = nb_of_birds - 1
    else:
        ar[n] = 1
        nb_of_birds = nb_of_birds + 1
    
    birds.append(nb_of_birds)

print(birds)
```
